```python
data.drop(columns = ['Organization Group Code', 'Job Family Code', 'Job Code', 'Year Type',
       'Year', 'Organization Group', 'Department',
       'Union Code', 'Union', 'Job Family', 'Employee Identifier',
       'Salaries', 'Overtime', 'Other Salaries', 'Total Salary', 'Retirement',
       'Health and Dental', 'Other Benefits', 'Total Benefits'], inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(data)
data = pd.DataFrame(min_max_scaler.transform(data))
data.columns = ['Department Code', 'Job', 'Total Compensation']
data.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data)
```

```python
kmeans.cluster_centers_
```

```python
y = kmeans.labels_
print(y)
```

```python
y.shape
```

```python
sns.countplot(x = kmeans.labels_, palette = 'Greens')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data['Job'].values,data['Total Compensation'].values, c=kmeans.labels_)
centers = np.array(kmeans.cluster_centers_)
plt.scatter(centers[:,0],centers[:,1],marker = 'x',color ='k')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(data['Department Code'].values,data['Job'].values, c=kmeans.labels_)
centers = np.array(kmeans.cluster_centers_)
plt.scatter(centers[:,0],centers[:,1],marker = 'x',color ='k')
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(data['Department Code'].values,data['Total Compensation'].values, c=kmeans.labels_)
centers = np.array(kmeans.cluster_centers_)
plt.scatter(centers[:,0],centers[:,1],marker = 'x',color ='k')
plt.show()
```
